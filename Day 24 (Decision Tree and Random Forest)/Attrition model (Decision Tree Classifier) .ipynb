## Convert Categorical in Numerical (OneHotEncoding)

```python
# Select Categorical columns only
cat_columns = df.select_dtypes(include='object')
cat_columns.shape
```

```python
# Number of unique features in all categorical columns
cat_columns.nunique()
```

__Here Number of categories are less than 10 for each columns. So we can directly apply OneHotEncoding.__

```python
cat_var = cat_columns.columns
df_one_hot = pd.get_dummies(df[cat_var],drop_first=True)
```

```python
df = pd.concat([df,df_one_hot], axis=1)
df.drop(df[cat_var], axis=1, inplace=True)
df.head()
```

## Built The Model (Decision Tree)

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import statsmodels.api as sm
```

```python
train, test = train_test_split(df, test_size= 0.3, random_state = 0)
dt = DecisionTreeClassifier()
```

```python
train.Attrition.hist(bins= 3)
```

```python
train.Attrition.value_counts()
```

```python
test.Attrition.value_counts()
```

```python
x_train = train.drop(['Attrition'], axis=1)
x_test = test.drop(['Attrition'], axis=1)
```

```python
y_train = train['Attrition']
y_test = test['Attrition']
```

```python
dt.fit(x_train, y_train)
```

```python
predict = dt.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
cl_rep =classification_report(y_test, predict)
cl_rep
```

```python
cm = confusion_matrix(y_test,predict)
score = accuracy_score(y_test,predict)
print('Accuracy score for Decision Tree Model',score)
print('Confusion Matrix: \n',cm)
```

```python
dt.score(x_train, y_train)
```

```python

from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import make_scorer, r2_score


cv = KFold(n_splits = 3, shuffle=True, random_state = 5)
r2 = make_scorer(r2_score)
r2_val_score = cross_val_score(dt, x_train, y_train, cv=cv, scoring = r2)
score = r2_val_score.mean()
score
```

__So, the accuracy for Decision Tree Model is quite good.__

__We can Increase our accuracy by apply RandomForest Model.__

__By Random Forest Classifier model we can identify the most significant features for our model.__

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=10, random_state= 0)
rf.fit(x_train, y_train)
```

```python
predict =rf.predict(x_test)

cm = confusion_matrix(y_test,predict)
score = accuracy_score(y_test,predict)
print('Accuracy score for Decision Tree Model',score)
print('Confusion Matrix: \n',cm)
```

```python
Rf = RandomForestClassifier(n_estimators=1000, max_features=2, oob_score=True)
Rf.fit(x_train, y_train)
```

```python
predict = Rf.predict(x_test)
```

```python
cm = confusion_matrix(y_test,predict)
score = accuracy_score(y_test,predict)
print('Accuracy score for Decision Tree Model',score)
print('Confusion Matrix: \n',cm)
```

```python
print(Rf.oob_score_)
```

```python
for i,k in zip(x_train.columns, Rf.feature_importances_):
    print(i,'\t \t: ',k)
```

__Now by using most significant features we can increase our accuracy score of Decision tree model, but our accuracy is already 98%.__

__So, we don't need to apply Decision tree model on most significant features only.__
