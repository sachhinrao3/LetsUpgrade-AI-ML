## Built a Model for Naive Bias theorem

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score
```

__Function to train model for Naive Bias theorem and get Accuracy score with Confusion matrix.__

```python
def model(train,test):
    x_train, x_test, y_train, y_test = train_test_split(train, test, test_size= 0.3, random_state =0)
    gaussian_nb = GaussianNB()
    predict = gaussian_nb.fit(x_train, y_train).predict(x_test)
    cnfusn_mat = confusion_matrix (predict, y_test)
    acuracy_score = accuracy_score(predict, y_test)
    print('Accuracy Score for our model is:', acuracy_score)
    print('Confusion matrix for our model is: \n', cnfusn_mat)
```
